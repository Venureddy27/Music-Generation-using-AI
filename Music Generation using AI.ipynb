```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
import numpy as np

# Load the CSV file
df = pd.read_csv(r'C:\Users\panna\OneDrive\Desktop\features_30_sec.csv')

# Select only the columns that contain numerical features
features_df = df.select_dtypes(include=['int64', 'float64'])

# Remove rows with None values
features_df = features_df.dropna()

# Normalize the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features_df)

# Reshape the input data to match the expected shape of the LSTM layer
features_scaled = features_scaled.reshape(features_scaled.shape[0], features_scaled.shape[1], 1)

# Create the input layer
# Adjusted input shape to (features_scaled.shape[1], 1)
input_layer = Input(shape=(features_scaled.shape[1], 1))

# Create the LSTM layer
lstm_layer = LSTM(128)(input_layer)

# Create the dense layers
dense_layer1 = Dense(128, activation='relu')(lstm_layer)
dense_layer2 = Dense(features_scaled.shape[1], activation='sigmoid')(dense_layer1)  # Output layer with the same number of features

# Create the model
model = Model(inputs=input_layer, outputs=dense_layer2)

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')


# Train the model
# Now pass features_scaled with the correct shape
model.fit(features_scaled, features_scaled, epochs=100, batch_size=32)

# Generate music
generated_music = model.predict(features_scaled)
```
